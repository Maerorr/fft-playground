```python

# Importing the required libraries
import numpy as np
import matplotlib.pyplot as plt

def bandpass(f, dbpoct, current_f):
    if current_f < f:
        return -( f / current_f) * dbpoct
    else:
        return -(current_f / f) * dbpoct


# Parameters
fs = 44100           # Sampling rate
N = 1024             # FFT size
f_c = 1000           # Center frequency (Hz)
f_step = fs / N      # Frequency resolution
dbpoct = 24          # dB per octave

freqs = np.fft.fftfreq(N, 1 / fs)
x = np.ones(N)       # Input signal 

for i in range(N):
    x[i] = x[i] * bandpass(f_c, dbpoct, freqs[i])

plt.figure(figsize=(12, 6))
plt.plot(freqs[:N // 2], x[:N//2])
plt.title("Original Spectrum")
plt.xlabel("Frequency (Hz)")
plt.ylabel("Magnitude (dB)")
plt.grid()
plt.ylim(-60, 60)


plt.tight_layout()
plt.show()
```
