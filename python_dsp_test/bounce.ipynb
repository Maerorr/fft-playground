```python
import math
import numpy as np
import matplotlib.pyplot as plt

# write a function that takes x and returns easeOutBounce(x) for x in [0, 1]
# easeOutBounce(x) is a function that starts at 1 and goes to 0

def easeOutBounce(x):
    if x < 1/2.75:
        return 7.5625*x*x
    elif x < 2/2.75:
        x -= 1.5/2.75
        return 7.5625*x*x + 0.75
    elif x < 2.5/2.75:
        x -= 2.25/2.75
        return 7.5625*x*x + 0.9375
    else:
        x -= 2.625/2.75
        return 7.5625*x*x + 0.984375

time = np.linspace(0, 1, 1000)


# plot the function
plt.plot(time, [easeOutBounce(x) for x in time])
plt.show()

```
